# Analyse Exploratoire : Détection de Fraude Bancaire
---
Dans cette étude, j'analyse un jeu de données de transactions par carte bancaire pour identifier les caractéristiques des fraudes. Mon objectif est de comprendre comment les transactions frauduleuses se distinguent des transactions légitimes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# J'ajuste le style de mes graphiques
sns.set_theme(style='whitegrid')

# Je charge le dataset
df = pd.read_csv('../data/creditcard.csv')

# J'affiche les dimensions du dataset
print(f'Dimensions du dataset : {df.shape}')
df.head()
```

## 1. Analyse du déséquilibre des classes
Je calcule ici le taux exact de fraude pour mesurer la difficulté du problème.

```python
fraude_count = df['Class'].value_counts()
taux_fraude = (fraude_count[1] / len(df)) * 100

print(f'Nombre de transactions normales : {fraude_count[0]}')
print(f'Nombre de fraudes : {fraude_count[1]}')
print(f'Taux de fraude : {taux_fraude:.4f}%')

# Visualisation du déséquilibre
plt.figure(figsize=(8, 6))
sns.countplot(x='Class', data=df, palette='magma')
plt.title('Déséquilibre des classes (0: Légitime, 1: Fraude)')
plt.yscale('log')
plt.show()
```

## 2. Distribution du montant des transactions
Je compare les montants pour voir si les fraudeurs ciblent des sommes spécifiques.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

sns.histplot(df[df['Class'] == 1]['Amount'], bins=30, ax=ax1, color='red', kde=True)
ax1.set_title('Distribution des montants - FRAUDES')

sns.histplot(df[df['Class'] == 0]['Amount'], bins=30, ax=ax2, color='blue', kde=True)
ax2.set_title('Distribution des montants - LÉGITIMES')
ax2.set_xlim(0, 2000)

plt.show()
```

## Conclusions Business
1. **Rareté extrême** : Avec seulement 0.17% de fraudes, je constate que mon futur modèle devra être extrêmement précis.
2. **Comportement des montants** : Je remarque que les fraudes sont souvent de petits montants, ce qui suggère des tentatives de passer inaperçu.
3. **Nécessité de robustesse** : Le déséquilibre m'impose d'utiliser des techniques de pondération lors de la modélisation.
